```
# ================================================================================
# CELL 6: Verify Connections (Version-Agnostic)
# ================================================================================
print("ğŸ” Verifying connections...\n")

def get_tools_for_entity(entity_id):
    """Version-agnostic tool getter"""
    if USE_TOOLSET:
        # Newer version
        return toolset.get_tools(
            actions=[
                Action.GOOGLECALENDAR_FIND_FREE_SLOTS,
                Action.GOOGLECALENDAR_CREATE_EVENT,
                Action.GMAIL_CREATE_EMAIL_DRAFT
            ],
            entity_id=entity_id
        )
    else:
        # Older version
        return composio.tools.get(
            user_id=entity_id,
            tools=[
                "GOOGLECALENDAR_FIND_FREE_SLOTS",
                "GOOGLECALENDAR_CREATE_EVENT",
                "GMAIL_CREATE_EMAIL_DRAFT"
            ]
        )

# Test Account 1
try:
    tools1 = get_tools_for_entity("HR_AREEB_HUBSPOT_567")
    print(f"âœ… Account 1: Got {len(tools1)} tools - CONNECTED!")
except Exception as e:
    print(f"âŒ Account 1: {e}")

# Test Account 2
try:
    tools2 = get_tools_for_entity("HR_AREEB_GENAI_567")
    print(f"âœ… Account 2: Got {len(tools2)} tools - CONNECTED!")
except Exception as e:
    print(f"âŒ Account 2: {e}")

print("\nğŸ‰ Ready for testing!")
```

```
# ================================================================================
# CELL 7 (FINAL FIX): Multi-HR Scheduling Function
# ================================================================================
from datetime import datetime

def schedule_interview(entity_id, account_name, candidate_name, candidate_email):
    """Schedule interview with proper datetime formatting"""
    print(f"\n{'='*70}")
    print(f"ğŸ“… SCHEDULING: {account_name.upper()}")
    print(f"{'='*70}")
    print(f"Entity: {entity_id}")
    print(f"Candidate: {candidate_name}\n")

    # Get tools
    all_tools = get_tools_for_entity(entity_id)

    # Identify tools by their name/schema instead of position
    find_slots = None
    create_event = None
    send_email = None

    for tool in all_tools:
        tool_name = tool.name.lower()
        if 'find' in tool_name and 'slot' in tool_name:
            find_slots = tool
        elif 'create' in tool_name and 'event' in tool_name:
            create_event = tool
        elif 'email' in tool_name or 'draft' in tool_name:
            send_email = tool

    if not all([find_slots, create_event, send_email]):
        print(f"âŒ Missing tools!")
        print(f"   Find slots: {find_slots is not None}")
        print(f"   Create event: {create_event is not None}")
        print(f"   Send email: {send_email is not None}")
        return None

    print(f"âœ… Tools identified correctly")

    # 1. Find free slots
    print("ğŸ” Finding free slots...")
    slots_result = find_slots.invoke({
        "date_range_start": "2025-12-14T00:00:00",
        "date_range_end": "2025-12-14T23:59:59",
        "timezone": "UTC",
        "calendar_ids": ["primary"]
    })

    # Handle different response structures
    if not slots_result.get('successful'):
        print(f"âŒ Failed to find slots: {slots_result.get('error')}")
        return None

    # Try different data structures
    try:
        if 'data' in slots_result and 'calendars' in slots_result['data']:
            free_slots = slots_result["data"]["calendars"]["primary"]["free"]
        elif 'data' in slots_result and 'free' in slots_result['data']:
            free_slots = slots_result["data"]["free"]
        elif 'free' in slots_result:
            free_slots = slots_result["free"]
        else:
            print(f"âŒ Unexpected response structure: {slots_result}")
            return None
    except Exception as e:
        print(f"âŒ Error parsing slots: {e}")
        print(f"Response was: {slots_result}")
        return None

    print(f"âœ… Found {len(free_slots)} slots")

    if not free_slots:
        print("âŒ No slots available")
        return None

    # Get first slot and fix datetime format
    slot = free_slots[0]
    start_str = slot["start"]
    if start_str.endswith("Z"):
        start_str = start_str.replace("Z", "+00:00")

    start_dt = datetime.fromisoformat(start_str)
    start_time = start_dt.strftime("%Y-%m-%dT%H:%M:%S")  # Remove microseconds

    print(f"ğŸ“… Slot: {start_time}")

    # 2. Create event
    print("\nğŸ“ Creating calendar event...")
    event_result = create_event.invoke({
        "start_datetime": start_time,
        "event_duration_hour": 1,
        "event_duration_minutes": 0,
        "summary": f"[{account_name}] Interview: {candidate_name}",
        "description": f"HR: {entity_id}\nCandidate: {candidate_name}\nEmail: {candidate_email}",
        "attendees": [candidate_email],
        "calendar_id": "primary",
        "timezone": "UTC"
    })

    if event_result.get('successful'):
        print("âœ… Event created!")
        if 'data' in event_result:
            event_id = event_result['data'].get('id', 'N/A')
            print(f"   Event ID: {event_id}")
    else:
        print(f"âŒ Event failed: {event_result.get('error')}")

    # 3. Send email - FIXED PARAMETER NAME
    print("\nğŸ“§ Creating email draft...")
    email_result = send_email.invoke({
        "recipient_email": candidate_email,  # âœ… Changed from "to" to "recipient_email"
        "subject": f"Interview Scheduled - {candidate_name}",
        "body": f"""Hello {candidate_name},

Your interview has been scheduled for {start_time} UTC.

Interviewer: {account_name}
Duration: 1 hour

Best regards,
HR Team"""
    })

    if email_result.get('successful'):
        print("âœ… Email draft created!")
    else:
        print(f"âŒ Email failed: {email_result.get('error')}")

    print(f"{'='*70}\n")

    return {
        'event': event_result,
        'email': email_result,
        'time': start_time
    }
```
